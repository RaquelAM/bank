```python
data[['Unnamed: 0', 'age','MonthlyIncome','DebtRatio']]
```

```python
remaining_money = data.MonthlyIncome * (1 - data.DebtRatio)
```

```python
data['remaining_money'] = remaining_money
```

```python
remaining_money_table = data[['Unnamed: 0', 'age','remaining_money']]
```

```python
remaining_money_table
```

```python
remaining_money_table = remaining_money_table.dropna(axis=0, how='any')
remaining_money_table
```

```python
remaining_money_table = remaining_money_table[~(remaining_money_table['remaining_money']<= 0)]  
```

```python
remaining_money_table.sort_values(by='remaining_money')
```

```python
age_remaining_money = pd.DataFrame(columns={'money', 'age'})
```

```python
age_remaining_money
```

```python
for x in range(21,104):
    mean = remaining_money_table.loc[remaining_money_table['age'] == x]["remaining_money"].mean()
    age_remaining_money = age_remaining_money.append(pd.DataFrame([[mean, x]], columns={'money','age'}))
```

```python
age_remaining_money
```

```python
age_remaining_money = age_remaining_money.dropna(axis=0, how='any')
age_remaining_money
```

Edad vs dinero sobrante al mes, despus de pagar deudas:

```python
age_remaining_money.plot(x='age', y='money',figsize=(20,10))
```

```python
age_remaining_money.sort_values(by='money')
```
